# Markovitz Portfolio Construction

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
assets = ['PG','^GSPC']
pf_data = pd.DataFrame()
for asset in assets:
    pf_data[asset] = wb.DataReader(asset,data_source='yahoo',start='2010-1-1')['Adj Close']
```

```python
pf_data.tail()
```

```python
((pf_data / pf_data.iloc[0])*100).plot(figsize=(16,6))
```

```python
log_returns = np.log(pf_data/pf_data.shift(1))
log_returns.tail()
```

```python
return_mean_y = log_returns.mean()*250
return_mean_y
```

```python
return_cov_y = log_returns.cov()*250
return_cov
```

```python
return_corr = log_returns.corr()*250
return_corr
```

```python
weights = np.random.random(2)
weights /= np.sum(weights)
np.sum(weights)
```

```python
# Expected Portfolio Returns
pfolio_expret = np.dot(weights.T,return_mean_y)
pfolio_expret
```

```python
# Portfolio variance
pfolio_var = (weights.T @ return_cov_y) @ weights
pfolio_var
```

```python
# Portfolio volatility
np.power(pfolio_var,0.5)
```

```python
# Weights combinations
pfolio_returns = []
pfolio_volatilities = []
weights_arr = []
for pg_weight in np.linspace(0,1,1000):
    sp_weight = 1 - pg_weight
    weights_arr.append(pg_weight)
    weights = np.array([pg_weight,sp_weight])
    pfolio_return = weights.T @ return_mean_y
    pfolio_returns.append(pfolio_return)
    pfolio_var = (weights.T @ return_cov) @ weights
    pfolio_volatilities.append(np.sqrt(pfolio_var))

pfolio_returns = np.array(pfolio_returns)
pfolio_volatilities = np.array(pfolio_volatilities)
weights_arr = np.array(weights_arr)
```

```python
markov_df = pd.DataFrame()
markov_df['Exp Return'] = pfolio_returns
markov_df['Volatility'] = pfolio_volatilities
markov_df.head()
```

```python
plt.figure(figsize=(10,7))
plt.title('Markov Efficient Frontier (Color = P&G Weight)')
plt.scatter(x=pfolio_volatilies,y=pfolio_returns,c=weights_arr)
plt.colorbar()
plt.xlabel('Volatility')
plt.ylabel('Return')
```

```python
plt.figure(figsize=(10,7))
plt.title('P&G Weight vs. Return (Color = Volatility)')
plt.scatter(x=weights_arr,y=pfolio_returns,
            c=pfolio_volatilities,cmap='viridis')
plt.xlabel('PG weight')
plt.ylabel('Portfolio Returns')
plt.colorbar()
```

## CAPM

```python
# Beta for last five years
tickers = ['PG','^GSPC']
data = pd.DataFrame()
for t in tickers:
    data[t] = wb.DataReader(t,data_source='yahoo',
                            start='2015-1-1', end='2019-12-31')['Adj Close']
```

```python
sec_returns = np.log(data / data.shift(1))
```

```python
cov = sec_returns.cov()*250
cov
```

```python
cov_market = cov.iloc[0,1]
cov_market
```

```python
market_var = sec_returns['^GSPC'].var()*250
market_var
```

```python
PG_beta = cov_market / market_var
PG_beta
```

PG is a <b>defensive</b> stock.

```python
avg_returns_y = sec_returns.mean()*250
avg_returns_y
```

```python
# Bloomberg 10 Year US Treasury Bond Yield (2020-3-29)
rf_rate = 0.67/100
rf_rate
```

```python
PG_ret = rf_rate + PG_beta*0.05
PG_ret
```

```python
# Sharpe Ratio
PG_Sharpe = (PG_ret - rf_rate) / (sec_returns['PG'].std()*(250**0.5))
PG_Sharpe
```
